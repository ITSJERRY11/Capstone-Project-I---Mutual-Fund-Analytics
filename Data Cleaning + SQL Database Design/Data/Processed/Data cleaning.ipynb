#Cleanning the data "Nav_history.CSV"

```python
# 1. Load the data
df_nav = pd.read_csv(r'C:\Users\TEJ\OneDrive\Desktop\Project\Data Cleaning + SQL Database Design\Datasets\02_nav_history.csv')
df_nav 
```

```python
# 2. Format Dates
df_nav['date'] = pd.to_datetime(df_nav['date'])
df_nav['date']
```

```python
# 3. Sort the data
df_nav = df_nav.sort_values(['amfi_code', 'date'])
df_nav
```

```python
# 4. Impute Missing Values (Forward-fill by amfi_code)
df_nav['nav'] = df_nav.groupby('amfi_code')['nav'].ffill()
df_nav
```

```python
# 5. Duplicate value removal
df_nav = df_nav.drop_duplicates()
df_nav
```

```python
# 6. Validate (Ensure all NAV > 0)
df_nav = df_nav[df_nav['nav'] > 0]
df_nav
```

```python
print(f"Cleaned NAV History shape: {df_nav.shape}")
```

#CLEANING INVESTOR-TRANSACTIONS CSV

```python
# 1. Load the data
df_trans = pd.read_csv(r'C:\Users\TEJ\OneDrive\Desktop\Project\Data Cleaning + SQL Database Design\Datasets\08_investor_transactions.csv')
df_trans
```

```python
# 2. Standardize Types (Strips whitespace and Capitalizes the first letter)
df_trans['transaction_type'] = df_trans['transaction_type'].astype(str).str.strip().str.title()
df_trans
```

```python
# 3. Validate Amounts
df_trans = df_trans[df_trans['amount_inr'] > 0]
df_trans
```

```python
# 4. Format Dates
df_trans['transaction_date'] = pd.to_datetime(df_trans['transaction_date'])
df_trans['transaction_date']
```

```python
# 5. Verify KYC Status
allowed_kyc = ['Verified', 'Pending', 'Rejected']
df_trans = df_trans[df_trans['kyc_status'].isin(allowed_kyc)]
df_trans
```

```python
print(f"Cleaned Transactions shape: {df_trans.shape}")
```

#CLEANING SCHEME-PERFORMANCE CSV

```python
# 1. Load the data
df_perf = pd.read_csv(r'C:\Users\TEJ\OneDrive\Desktop\Project\Data Cleaning + SQL Database Design\Datasets\07_scheme_performance.csv')
df_perf.head(5)
```

```python
# 2. Numeric Validation
# Replace 'return_1y', 'return_3y', etc., with your actual column names
return_columns = ['return_1y', 'return_3y', 'return_5y'] 

for col in return_columns:
    if col in df_perf.columns:
        df_perf[col] = pd.to_numeric(df_perf[col], errors='coerce')

# Drop rows where any of the return columns became NaN due to coercion
existing_cols = [col for col in return_columns if col in df_perf.columns]
df_perf = df_perf.dropna(subset=existing_cols)

df_perf.head(6)
```

```python
# 3. Expense Ratio Check
df_perf = df_perf[(df_perf['expense_ratio_pct'] >= 0.1) & (df_perf['expense_ratio_pct'] <= 2.5)]
df_perf.head(6)
print(f"Cleaned Performance shape: {df_perf.shape}")
```
